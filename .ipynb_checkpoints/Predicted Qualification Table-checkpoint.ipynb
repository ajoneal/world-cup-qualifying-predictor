```python
import pandas as pd
import numpy as np
```

```python
current_CONCACAF_table = pd.read_csv('./data/current_table_12.14.2021.csv')
current_CONCACAF_table
```

```python
predictions_dict = {'team_name': ['Canada', 'United States', 'Mexico', 'Panama', 'Costa Rica', 'Jamaica', 'El Salvador', 'Honduras'], 
'matches_played': [0, 0, 0, 0, 0, 0, 0, 0], 'wins': [0, 0, 0, 0, 0, 0, 0, 0], 'draws': [0, 0, 0, 0, 0, 0, 0, 0],
'losses': [0, 0, 0, 0, 0, 0, 0, 0], 'goals_for': [0, 0, 0, 0, 0, 0, 0, 0], 'goals_against': [0, 0, 0, 0, 0, 0, 0, 0], 
'goal_differential': [0, 0, 0, 0, 0, 0, 0, 0], 'points': [0, 0, 0, 0, 0, 0, 0, 0]}
```

```python
predicted_table = pd.DataFrame(predictions_dict)
predicted_table
```

```python
qualifying_fixtures = pd.read_csv('./data/qualifier_predictions.csv')
qualifying_fixtures
```

```python
qualifying_fixtures_dict = qualifying_fixtures.to_dict('records')
qualifying_fixtures_dict[0]
```

```python
predicted_table_dict = predicted_table.to_dict('records')
predicted_table_dict[0]
```

```python
if qualifying_fixtures_dict[0]['home_team'] == predicted_table_dict[0]['team_name']:
    predicted_table_dict[0]['matches_played'] = predicted_table_dict[0]['matches_played'] + 1
    predicted_table_dict[0]['goals_for'] = predicted_table_dict[0]['goals_for'] + qualifying_fixtures_dict[0]['home_score']
    predicted_table_dict[0]['goals_against'] = predicted_table_dict[0]['goals_against'] + qualifying_fixtures_dict[0]['away_score']
    predicted_table_dict[0]['goal_differential'] = predicted_table_dict[0]['goal_differential'] + qualifying_fixtures_dict[0]['home_score'] - qualifying_fixtures_dict[0]['away_score']
    if qualifying_fixtures_dict[0]['home_score'] > qualifying_fixtures_dict[0]['away_score']:
        predicted_table_dict[0]['wins'] = predicted_table_dict[0]['wins'] + 1
        predicted_table_dict[0]['points'] = predicted_table_dict[0]['points'] + 3
    elif qualifying_fixtures_dict[0]['home_score'] == qualifying_fixtures_dict[0]['away_score']:
        predicted_table_dict[0]['draws'] = predicted_table_dict[0]['draws'] + 1
        predicted_table_dict[0]['points'] = predicted_table_dict[0]['points'] + 1
    else:
        predicted_table_dict[0]['losses'] = predicted_table_dict[0]['losses'] + 1
    
predicted_table_dict[0]
```

```python
for i in qualifying_fixtures_dict:
    for j in predicted_table_dict:
        if i['home_team'] == j['team_name']:
            j['matches_played'] = j['matches_played'] + 1
            j['goals_for'] = j['goals_for'] + i['home_score']
            j['goals_against'] = j['goals_against'] + i['away_score']
            j['goal_differential'] = j['goal_differential'] + i['home_score'] - i['away_score']
            if i['home_score'] > i['away_score']:
                j['wins'] = j['wins'] + 1
                j['points'] = j['points'] + 3
            elif i['home_score'] == i['away_score']:
                j['draws'] = j['draws'] + 1
                j['points'] = j['points'] + 1
            elif i['home_score'] < i['away_score']:
                j['losses'] = j['losses'] + 1
    
pd.DataFrame(predicted_table_dict)
```

```python
for i in qualifying_fixtures_dict:
    for j in predicted_table_dict:
        if i['away_team'] == j['team_name']:
            j['matches_played'] = j['matches_played'] + 1
            j['goals_for'] = j['goals_for'] + i['away_score']
            j['goals_against'] = j['goals_against'] + i['home_score']
            j['goal_differential'] = j['goal_differential'] + i['away_score'] - i['home_score']
            if i['away_score'] > i['home_score']:
                j['wins'] = j['wins'] + 1
                j['points'] = j['points'] + 3
            elif i['away_score'] == i['home_score']:
                j['draws'] = j['draws'] + 1
                j['points'] = j['points'] + 1
            elif i['away_score'] < i['home_score']:
                j['losses'] = j['losses'] + 1
    
pd.DataFrame(predicted_table_dict)
```
